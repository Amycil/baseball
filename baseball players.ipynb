```python
import pandas as pd

baseball_data = pd.read_csv('baseball_players.csv')
baseball_data.head()
```

```python
#EDA
baseball_data = baseball_data.dropna()
baseball_data.isnull().sum()#check for null values

#dtypes
baseball_data.dtypes
baseball_data.info()
baseball_data.describe()
```

```python
#Rename Height and Weight columns

baseball_data = baseball_data.rename({'Height(inches)' : 'Height', 'Weight(pounds)':'Weight'}, axis=1)
baseball_data.head()
```

```python
import matplotlib.pyplot as plt

plt.subplot(2, 2, 1)
plt.hist(baseball_data.Height, bins=10)
plt.title("Height of Baseball Players")
plt.show()

plt.subplot(2, 3, 2)
plt.hist(baseball_data.Weight, bins=10)
plt.show()
```

```python
#relationship between age and weight
import numpy as np
baseball_data['Age'] = round(baseball_data['Age'], 0)# round the age to a whole num

plt.scatter(baseball_data.Age, baseball_data.Weight)
z = np.polyfit(baseball_data.Age, baseball_data.Weight, 1)
line_function = np.poly1d(z)
plt.plot(baseball_data.Age, line_function(baseball_data.Age), "r-")

plt.show()
```

```python
#weight and height
plt.scatter(baseball_data.Weight, baseball_data.Height)
z = np.polyfit(baseball_data.Weight, baseball_data.Height, 1)
line_function = np.poly1d(z)
plt.plot(baseball_data.Weight, line_function(baseball_data.Weight), "r-")

plt.show()
```

```python
plt.scatter(baseball_data.Age, baseball_data.Height)
z = np.polyfit(baseball_data.Age, baseball_data.Height, 1)
line_function = np.poly1d(z)
plt.plot(baseball_data.Age, line_function(baseball_data.Age), "r-")

plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,6))
ax = sns.boxplot(data=baseball_data, x=baseball_data.Position, y=baseball_data.Height)
sns.set_style('darkgrid')
plt.xticks(rotation=40)
plt.title("Height Of Baseball Players per Position")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=baseball_data, x=baseball_data.Position, y=baseball_data.Weight)
plt.xticks(rotation=40)
plt.title("Weight of Baseball Players per Position")
plt.show()
```
